```python
from sklearn.datasets import make_regression
import numpy as np

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### **Gradient Descent Intuation**

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80,random_state=13)
```

```python
plt.figure(figsize = (4,4))
plt.scatter(x = X, y = y)
plt.show()
```

```python
lr = LinearRegression()
```

```python
lr.fit(X, y)
```

#### Slope and Intercept using sklearn class

```python
m = lr.coef_[0]
b = lr.intercept_
print(m, b)
```

#### Best Fit Line

```python
plt.figure(figsize = (4,4))
plt.scatter(x = X, y = y)
plt.plot(X, (m*X + b), color = 'red')
plt.show()
```

#### Lets apply Gradient Descent
- Starting random values of m = 1 and b = 0

```python
m = 1
b = 0
learning_rate = 0.1
```

```python
y_pred = ((m * X) + b).reshape(4)
```

```python
plt.figure(figsize = (4,4))
plt.scatter(x = X, y = y)
plt.plot(X, lr.predict(X), color = 'red')
plt.plot(X, y_pred, color = 'black')
plt.show()
```

### 1st Iteration

```python
y_hat = m * X.reshape(4) + b
```

#### Calculating Intercept

```python

slope = -2 * np.sum(y - y_hat)
print(slope)

b_new = b - (learning_rate * slope)
print(b_new)
```

#### Calculating Slope

```python
slope = -2 * np.dot(y - y_hat, X)
print(slope)

m_new = m - learning_rate * slope
print(m_new)
```

```python
plt.figure(figsize = (4,4))
plt.scatter(x = X, y = y)
plt.plot(X, lr.predict(X), color = 'red')
plt.plot(X, y_pred, color = 'black')
plt.plot(X, (m_new * X) + b_new, color = 'green')
plt.show()
```

### 2nd Iteration

```python
y_hat = m_new * X.reshape(4) + b_new
```

#### Finding Intercept using b_new

```python
slope = -2 * np.sum(y - y_hat)
print(slope)

second_b_new = b_new - (learning_rate * slope)
print(second_b_new)
```

#### Finding slope using m_new

```python
slope = -2 * np.dot(y - y_hat, X)
print(slope)

second_m_new = m_new - (learning_rate * slope)
print(m_new)
```

```python
plt.figure(figsize = (4,4))
plt.scatter(x = X, y = y)
plt.plot(X, lr.predict(X), color = 'red')
plt.plot(X, y_pred, color = 'black')
plt.plot(X, (m_new * X) + b_new, color = 'green')
plt.plot(X, (second_m_new * X) + second_b_new, color = 'blue')
plt.show()
```

### A complete visualization using Inetating No. of Epochs times

```python
b = 0
m = 1
epochs = 250
learning_rate = 0.01

for i in range(epochs):

    y_hat = (m * X.reshape(4)) + b

    slope = -2 * np.sum(y - y_hat)
    b = b - (learning_rate * slope)

    slope = -2 * np.dot((y - y_hat), X.reshape(4))
    m = m - (learning_rate * slope)

    plt.plot(X, (m * X) + b)

plt.scatter(X, y)
plt.show()
```
